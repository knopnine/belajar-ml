```python
import pandas as pd

# define dataset path
# dataset source https://www.kaggle.com/datasets/mohitkumar282/used-car-dataset
path = r"dataset\used_car_dataset.csv"
car_data = pd.read_csv(path)

#cek 5 data teratas
car_data.head(5)
```

```python

#cek kolom yang ada di dataset
car_data.columns
```

```python
#drop unrelated columns
car_data = car_data.drop("AdditionInfo", axis=1)
```

```python
#cek info detail dari tiap kolom
print(car_data.info())
```

```python
#check data with empty value
car_data[pd.isnull(car_data.kmDriven)]
```

```python
#drop empy data from dataset
car_data.dropna(subset='kmDriven').reset_index(drop=True)
```

```python
#since i live in Indonesia, i want to know the price of car in this dataset in my own currency
#create new column by converting askprice type from object to integer, deleting currency symbol and replace floating number
car_data['price_in_lakhs'] = car_data['AskPrice'].str.replace('₹', '').str.replace(',', '').astype('Int64')

#convert the price to rupiah (at the moment i write the code, 1 lakh = 188.292 rupiah)
car_data['price_in_rupiah'] = car_data['price_in_lakhs']*188.292
car_data['price_in_rupiah'] = car_data['price_in_rupiah'].astype("int64")
car_data.sort_values("price_in_rupiah", ascending=True)
```

```python
#converting kmdriven to int
car_data['car_driven_in_km'] = car_data['kmDriven'].str.replace('km', '').str.replace(',', '').astype(float)
car_data.head(5)
```
