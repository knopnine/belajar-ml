```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a dummy DataFrame
data = pd.DataFrame({
    'House ID': np.arange(1, 51),
    'Price': np.random.randint(100000, 1000000, size=50),
    'Square Meter': np.random.randint(50, 300, size=50),
    'Bedrooms': np.random.randint(1, 10, size=50),
    'Bathrooms': np.random.randint(1, 5, size=50),
    'Year Built': np.random.randint(1950, 2023, size=50)
})

# View the data
print(data)
```

```python
# Calculate central tendency
# Mean

mean_price = data['Price'].mean()
mean_square_meter = data['Square Meter'].mean()
mean_bedrooms = data['Bedrooms'].mean()
mean_bathrooms = data['Bathrooms'].mean()

print('Mean Price:', mean_price)
print('Mean Square Feet:', mean_square_meter)
print('Mean Bedrooms:', mean_bedrooms)
print('Mean Bathrooms:', mean_bathrooms)

# Median
median_price = data['Price'].median()
median_square_meter = data['Square Meter'].median()
median_bedrooms = data['Bedrooms'].median()
median_bathrooms = data['Bathrooms'].median()

print('Median Price:', median_price)
print('Median Square Feet:', median_square_meter)
print('Median Bedrooms:', median_bedrooms)
print('Median Bathrooms:', median_bathrooms)

# Mode
mode_price = data['Price'].mode()
mode_square_meter = data['Square Meter'].mode()
mode_bedrooms = data['Bedrooms'].mode()
mode_bathrooms = data['Bathrooms'].mode()

print('Mode Price:', mode_price)
print('Mode Square Meter:', mode_square_meter)
print('Mode Bedrooms:', mode_bedrooms)
print('Mode Bathrooms:', mode_bathrooms)
```

```python
# Calculate dispersion
# Range

range_price = data['Price'].max() - data['Price'].min()
range_square_meter = data['Square Meter'].max() - data['Square Meter'].min()
range_bedrooms = data['Bedrooms'].max() - data['Bedrooms'].min()
range_bathrooms = data['Bathrooms'].max() - data['Bathrooms'].min()

print('Range Price:', range_price)
print('Range Square Meter:', range_square_meter)
print('Range Bedrooms:', range_bedrooms)
print('Range Bathrooms:', range_bathrooms)
print('\n')

# Variance
variance_price = data['Price'].var()
variance_square_meter = data['Square Meter'].var()
variance_bedrooms = data['Bedrooms'].var()
variance_bathrooms = data['Bathrooms'].var()

print('Variance Price:', variance_price)
print('Variance Square Meter:', variance_square_meter)
print('Variance Bedrooms:', variance_bedrooms)
print('Variance Bathrooms:', variance_bathrooms)
print('\n')

# Standard Deviation
std_price = data['Price'].std()
std_square_meter = data['Square Meter'].std()
std_bedrooms = data['Bedrooms'].std()
std_bathrooms = data['Bathrooms'].std()

print('Standard Deviation Price:', std_price)
print('Standard Deviation Square Meter:', std_square_meter)
print('Standard Deviation Bedrooms:', std_bedrooms)
print('Standard Deviation Bathrooms:', std_bathrooms)
print('\n')

# interquartile range
Q1_price = data['Price'].quantile(0.25)
Q3_price = data['Price'].quantile(0.75)
IQR_price = Q3_price - Q1_price

Q1_square_feet = data['Square Meter'].quantile(0.25)
Q3_square_feet = data['Square Meter'].quantile(0.75)
IQR_square_feet = Q3_square_feet - Q1_square_feet

Q1_bedrooms = data['Bedrooms'].quantile(0.25)
Q3_bedrooms = data['Bedrooms'].quantile(0.75)
IQR_bedrooms = Q3_bedrooms - Q1_bedrooms

Q1_bathrooms = data['Bathrooms'].quantile(0.25)
Q3_bathrooms = data['Bathrooms'].quantile(0.75)
IQR_bathrooms = Q3_bathrooms - Q1_bathrooms

print('IQR Price:', IQR_price)
print('IQR Square Meter:', IQR_square_feet)
print('IQR Bedrooms:', IQR_bedrooms)
print('IQR Bathrooms:', IQR_bathrooms)
```

```python
# Visualize the data
# Histogram

sns.histplot(data['Price'], kde=True)
plt.title('Price Distribution')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()

sns.histplot(data['Square Meter'], kde=True)
plt.title('Square Feet Distribution')
plt.xlabel('Square Feet')
plt.ylabel('Frequency')
plt.show()

sns.histplot(data['Bedrooms'], kde=True)
plt.title('Bedrooms Distribution')
plt.xlabel('Bedrooms')
plt.ylabel('Frequency')
plt.show()

sns.histplot(data['Bathrooms'], kde=True)
plt.title('Bathrooms Distribution')
plt.xlabel('Bathrooms')
plt.ylabel('Frequency')
plt.show()

# Boxplot   
sns.boxplot(data['Price'])
plt.title('Price Distribution')
plt.xlabel('Price ($)')
plt.show()

sns.boxplot(data['Square Meter'])
plt.title('Square Meter Distribution')
plt.xlabel('Square Meter')
plt.show()

sns.boxplot(data['Bedrooms'])
plt.title('Bedrooms Distribution')
plt.xlabel('Bedrooms')
plt.show()
```
